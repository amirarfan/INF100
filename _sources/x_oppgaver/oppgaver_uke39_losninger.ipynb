```julia
# Oppgave 1: Funksjonsplott
function plot_functions()
    x = -10:0.1:10

    y1 = 1 ./ x
    plot(x, y1, linestyle=:dash, color=:blue, label="y = 1/x", grid=:both)
    xlabel!("x")
    ylabel!("y")
    title!("Funksjonsplott")

    y2 = sin.(x) .* cos.(x)
    plot!(x, y2, linestyle=:dot, color=:red, label="y = sin(x)cos(x)")

    y3 = 2 .* x.^2 .- 3 .* x .+ 1
    scatter!(x, y3, marker=:x, color=:red, label="y = 2x^2 - 3x + 1")
end
```

```julia
plot_functions()
```

```julia
function circle(h, k, r)
    t = range(0,2pi,100)
    h .+ r*sin.(t), k .+ r*cos.(t)
end 
# (h,k,r) er sentrum av sirkel og radius
plot(circle(0,0,5),aspect_ratio=:equal, seriestype = [:shape], c = :blue, lw =:2, fillalpha =:0.2) # x,y = 0,0 med radius 5
```

```julia
# Leminscate x og y plot
function leminscate()
    t = range(-pi/4,pi/4, 100)
    x = cos.(t) .* sqrt.(2 .* cos.(2 .* t))
    y = sin.(t) .* sqrt.(2 .* cos.(2 .* t))
    plot(x, y, color=:red)
end

leminscate()
```

#### c)

```julia
# Lograthmic spiral 
function log_spiral()
    t = range(0, 6pi, 100)
    k = 0.1
    x = exp.(k .* t) .* cos.(t)
    y = exp.(k .* t) .* sin.(t)
    plot(x, y, color=:green)
end

log_spiral()
```

```julia
# Oppgave 3: If-else forstÃ¥else og funksjoner
function days_in_month(month::String)
    if month == "Januar" # Kunne endret til Ã¥ bruke lowercase(month) == "januar" , som ville vÃ¦rt en bedre lÃ¸sning
        return 31
    elseif month == "Februar"
        return "28 eller 29"
    elseif month in ["Mars", "Mai", "Juli", "August", "Oktober", "Desember"]
        return 31
    elseif month in ["April", "Juni", "September", "November"]
        return 30
    else
        return "Ugyldig mÃ¥ned"
    end
end

function maximum_of_three(a, b, c)
    if a >= b && a >= c
        return a
    elseif b >= a && b >= c
        return b
    else
        return c
    end
end

function quiz_game()
    correct_answers = 0
    wrong_answers = 0

    println("SpÃ¸rsmÃ¥l 1: Er jorden flat? (Ja/Nei)")
    answer1 = readline()
    if lowercase(answer1) == "nei"
        correct_answers += 1
        println("Riktig!")
    else
        wrong_answers += 1
        println("Galt!")
    end

    println("Antall riktige svar: ", correct_answers)
    println("Antall gale svar: ", wrong_answers)
end
```
