## Oppgave 4

````julia
using Statistics 
temperaturer = [17, 20, 21, 19, 18, 23, 22]
gjennomsnitt = mean(temperaturer)
println("Gjennomsnittstemperatur: ", gjennomsnitt)

# Eller med sum og lengde

gjennomsnitt = sum(temperaturer)/length(temperaturer)
println("Gjennomsnittstemperatur (med sum og lengde): ", gjennomsnitt)


# Eller med en for-løkke

summen = 0
for i in temperaturer
    summen += i
end
gjennomsnitt = summen/length(temperaturer)
println("Gjennomsnittstemperatur (med for-løkke): ", gjennomsnitt)
````
