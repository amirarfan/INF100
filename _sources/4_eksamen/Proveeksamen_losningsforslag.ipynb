## Oppgave 1a

```julia
function kalkulator(a, b, metode)
    metode = lowercase(metode) # Gjør om til små bokstaver fordi det er mer robust 
    if metode == "addere"
        return a + b
    elseif metode == "subtrahere"
        return a - b
    elseif metode == "multiplisere"
        return a * b
    elseif metode == "dividere"
        return a / b
    else
        # return "Ugyldig metode" # Dette er en mulighet
        error("Ugyldig metode") # Dette er en annen mulighet, for ekstrapoeng
    end
end
```

## Oppgave 1b


```julia
using Random
Random.seed!(1234) # Setter seed for å få samme tall hver gang
original_A = rand(1:10, 1, 100)

A = copy(original_A) # Lager en kopi av original_A fordi jeg skal sammenligne den etterpå, dette er ikke en del av oppgaven for dividere

println(A)
summe = 0.0
for i in 1:length(A) # for i in 1:100 eller for i, elem in enumerate(A) 
    A[i] += 5 # Legger til 5 på hvert element
    summe = kalkulator(summe, A[i], "Addere")
end
println(A)
println(summe)
```

```julia
# Med elementvis operasjoner 
A = copy(original_A)
A .+= 5 # Legger til 5 på hvert element, hva skjer hvis vi skriver A += 5?
summe = sum(A)
println(A)
println(summe)
```

## Oppgave 2a

```julia
function kuppelareal(a, b)
    return a * b + (pi * a^2) / 8
end
```

## Oppgave 2b

```julia
TA = zeros(10, 10)
for a in 1:10
    for b in 1:10
        areal = kuppelareal(a, b)
        # TA[a,b] = areal > 150 ? 150 : areal # Setter maks areal til 150, ekstrapoeng oppgave, fancy løsning
        # TA[a,b] = min(areal, 150) # Setter maks areal til 150, enklere løsning
        # Minst elegant løsning, men gir også fullt utbytte av ekstrapoeng 
        if areal > 150
            TA[a,b] = 150
        else
            TA[a,b] = areal
        end
    end
end
display(TA)
```

## Oppgave 2c

```julia
# En løsning:
count = 0
for i in 1:10
    for j in 1:10
        if TA[i, j] >= 40
            count += 1
        end
    end
end
println(count)

# En annen løsning:
count = 0
for elem in TA
    if elem >= 40
        count += 1
    end
end
println(count)

# En avansert løsning:
count = sum(TA .>= 40) # Prøv å skriv TA .>= 40 i REPL
println(count)
```

## Oppgave 4

Funksjonen `hemmeligH2023` ser ut til å implementere biseksjonsmetoden, som er en numerisk teknikk for å finne nullpunktet av en kontinuerlig funksjon. En av forutsetningene for at denne metoden skal fungere er at fortegn må byttes i løpet av intervallet, og det er det Linje 8-10 sjekker.  På eksamen burde man beskrive funksjonen i mer detalj:


#### Funksjonsdeklarasjon og Parametere
- **Linje 1:** Definerer funksjonen `hemmeligH2023` med parametrene `f`, `a`, `b`, `k` og `j`.
  - `f`: En funksjon som vi ønsker å finne rotverdien til.
  - `a` og `b`: Start- og sluttpunktene til et intervall. Funksjonen antar at det finnes en rot innenfor dette intervallet.
  - `k`: En toleransegrense for når vi skal stoppe iterasjonen (standard satt til `1e-6`).
  - `j`: Maksimalt antall iterasjoner (standard satt til 1000).

#### Inngangsverifisering
- **Linjer 8-10:** Sjekker om funksjonen `f` har forskjellige fortegn ved `a` og `b`. Hvis `f(a) * f(b) > 0`, indikerer det at det kanskje ikke finnes en rot i intervallet, og funksjonen gir en feilmelding. En fornuftig feilmelding kunne være `"Ingen garanti for rot i intervallet [a, b]"`.

#### Iterativ Prosess
- **Linjer 12-24:** Utfører den iterative prosessen for å finne rotverdien.
  - **Linje 13:** Initialiserer en teller `i` for antall gjennomførte iterasjoner.
  - **Linje 14:** Beregner midtpunktet `c` mellom `a` og `b`.
  - **Linjer 15-16:** Sjekker om verdien av `f` ved `c` er innenfor toleransegrensen `k`, eller om intervallet `[a, b]` har blitt tilstrekkelig lite. Hvis en av disse betingelsene er oppfylt, returnerer funksjonen `c` som en tilnærming til rotverdien.
  - **Linjer 18-22:** Oppdaterer intervallet basert på hvor rotverdien antas å ligge. Hvis produktet av `f(a)` og `f(c)` er negativt, betyr det at roten ligger mellom `a` og `c`, og `b` settes til `c`. Hvis ikke, oppdateres `a` til `c`.
  - **Linje 23:** Øker telleren `i` med 1 for hver iterasjon.
- Hvis maksimalt antall iterasjoner `j` blir nådd uten at en tilstrekkelig nøyaktig rotverdi er funnet, avsluttes funksjonen uten en returverdi. Her kunne man lagt til en feilmelding slik som `"Maksimalt antall iterasjoner nådd uten konvergens."`

#### Sammendrag
`hemmeligH2023` er en implementasjon av bisection metoden for å finne approksimerte rotverdier av en kontinuerlig funksjon innenfor et gitt intervall. Den bruker en kombinasjon av en toleransegrense for funksjonsverdien og en begrensning på antall iterasjoner for å sikre at metoden konvergerer mot en løsning eller stopper hvis den ikke finner en passende løsning innenfor de gitte grensene.
