## Hva er Eksponentiell Vekst?

Forståelse av eksponentiell vekst forutsetter at man har fulgt med i kurset ECN102 hvor eksponentielle funksjoner har blitt gjennomgått.

La oss starte med grunnleggende: I eksponentiell vekst øker mengden proporsjonalt med sin nåværende verdi. Dette betyr at jo større befolkningen er, desto raskere vokser den. I matematiske termer kan vi uttrykke dette som:

$$
B(t) = B_0 \cdot e^{rt}
$$

hvor 

- $B(t)$ er mengden på et gitt tidspunkt $t$
- $B_0$ er mengden ved $t=0$ (initial befolkningen)
- $r$ er vekstraten
- $e$ er Eulers tall, ca. 2.71828 (2.7 Ibsen Ibsen)

> Lineær vekst derimot er når mengden øker med en konstant mengde per tidsenhet. For eksempel, hvis du har 1000 mennesker som flytter til Oslo hvert år, vil befolkningen øke med 1000 mennesker hvert år. Dette er lineær vekst.
> Dette er en viktig forskjell mellom lineær og eksponentiell vekst: I eksponentiell vekst øker mengden med en prosentandel av sin nåværende verdi, mens i lineær vekst øker mengden med en konstant mengde.







## Eksponentiell vekst i to former:


#### Eksponentialfunksjon med Eulers Tall:

En klassisk modell for eksponentiell vekst er gitt ved formelen:

$$B(t) = B_0 \cdot e^{rt}$$

der $e$ er Eulers tall (cirka 2.71828), $r$ er vekstraten, $t$ er tiden, og $B_0$ er initialverdien.

#### Eksponentialfunksjon med Basis 1 + p:

I mange praktiske anvendelser, spesielt i finans og demografi, er det vanlig å uttrykke eksponentiell vekst som:

$$B(t) = B_0 \cdot (1 + p)^t$$

der $p$ representerer vekstraten som en desimal (for eksempel, 5% vekst er uttrykt som 0.05), $t$ er antall tidsperioder, og $B_0$ er initialverdien.

Begge formene er eksponentielle fordi de beskriver situasjoner hvor noe vokser i et tempo som er proporsjonalt med dets nåværende størrelse.

#### Forholdet Mellom dem:

For å forstå sammenhengen mellom de to formene, kan vi se på Eulers formel for eksponentiell vekst:

$$e^{rt} = (e^r)^t$$

Hvis vi sammenligner dette med den andre formelen, kan vi se at $e^r$ tilsvarer $1 + p$. Dette betyr at for en gitt vekstrate $r$, kan vi finne et tilsvarende $p$ slik at:

$$e^r = 1 + p$$ 

eller omvendt, gitt en prosentraten $p$, kan vi finne en kontinuerlig vekstrate $r$ som:

$$r = \ln(1 + p)$$ 

Så når vi arbeider med dagens forelesning og anvender dette til jordens befolkning, er vi faktisk i ferd med å bruke en eksponentiell vekstmodell som er lettere å forstå og beregne for mange mennesker, siden den bruker en prosentraten vi kan relatere til direkte.

I begge tilfeller handler eksponentiell vekst om å ta noe til en potens, enten det er Eulers tall eller (1 + p). Dette potensforholdet er hva som definerer veksten som eksponentiell.

## Hvorfor er Eksponentiell Vekst Viktig?

Eksponentiell vekst er svært relevant i økonomi for en rekke fenomener. Her er noen eksempler på hvordan eksponentiell vekst opptrer innen økonomisk teori og praksis:

1. **Renters rente**: Dette er kanskje det mest direkte eksemplet på eksponentiell vekst i økonomi. Når man investerer penger til en fast rente, vil man ikke bare tjene penger på det opprinnelige beløpet (hovedstolen), men også på rentene som allerede er tjent. Dette fører til at verdien av investeringen vokser eksponentielt over tid, et fenomen som Albert Einstein angivelig kalte "verdens åttende underverk".

2. **Økonomisk vekst**: Mange modeller av økonomisk vekst, spesielt de på lang sikt, bruker eksponentiell vekst til å beskrive hvordan produksjonsutgangen i en økonomi kan vokse. For eksempel, hvis en økonomi vokser med en konstant prosentsats per år på grunn av faktorer som teknologisk innovasjon og kapitalakkumulering, vil den vise eksponentiell vekst.

3. **Populasjonsvekst**: Eksponentiell vekstmodeller brukes også til å beskrive befolkningsvekst i en økonomisk kontekst, hvor befolkningens størrelse har direkte innflytelse på arbeidsstyrken, markedene, og etterspørselen i en økonomi.

4. **Inflasjon**: Over tid kan prisnivået i en økonomi vokse eksponentielt, spesielt i perioder med høy inflasjon. Dette har betydelige konsekvenser for kjøpekraften til valutaen.

5. **Spredning av teknologi**: Adopsjon av ny teknologi kan ofte følge en eksponentiell kurve hvor bruk av teknologi raskt vokser etter at en kritisk masse er nådd.

6. **Finansielle markeder**: Aksjekurser og markedsverdier kan også oppleve eksponentiell vekst, spesielt i bull-markeder, når investorenes tillit og kapitalinnsprøytinger driver verdiene oppover.

7. **Gjeldsvekst**: Lån og nasjonal gjeld kan vokse eksponentielt, noe som kan lede til ustabilitet eller kriser hvis ikke gjeldsveksten kontrolleres forsvarlig.

Forståelse av eksponentiell vekst er derfor avgjørende for økonomer, investorer og politikkutformere for å kunne lage prognoser, vurdere bærekraften til økonomiske trender og utforme økonomisk politikk.

### Bestemme Vekstrate

```julia
# Bestemme vekstraten $p$ som opfyller likningen B[1] * (1+p)^(2020-1950) = B[15]
# Med newtons metode kan vi finne en tilnærming til $p$

# Definer funksjonen som skal ha nullpunkt
f(p) = B[1] * (1+p)^(2020-1950) - B[15]

# Definer den deriverte av funksjonen
df(p) = (2020-1950) * B[1] * (1+p)^(2020-1950-1)

# Definer startverdien
p0 = 0.01

# Definer toleransen
tol = 1e-6

# Definer maks antall iterasjoner
max_iter = 100

# Definerer newtonrhapson funksjonen
function newtonraphson(f, df, p0, tol, max_iter)
    p = p0
    for i in 1:max_iter
        p = p - f(p)/df(p)
        if abs(f(p)) < tol
            return p
        end
    end
    return p
end

# Kjører newtonrhapson funksjonen
p = newtonraphson(f, df, p0, tol, max_iter)
```

```julia
# Kunne vi gjort det med bisection?

function bisect(f, a, b, tol, max_iter)
    for i in 1:max_iter
        c = (a+b)/2
        if abs(f(c)) < tol
            return c
        end
        if f(a)*f(c) < 0
            b = c
        else
            a = c
        end
    end
    return c
end

p = bisect(f, 0, 10, tol, max_iter)
```

### Betydningen av likningen

Likningen vi stilte opp i deloppgaven

$B(1) \cdot (1+p)^{2020-1950} = B(15)$

representerer en matematisk modell for eksponensiell vekst, hvor $B(1)$ er befolkningsstørrelsen i begynnelsen av perioden (i år 1950), $B(15)$ er befolkningsstørrelsen ved slutten av perioden (i år 2020), og $p$ er den årlige vekstraten.

Eksponensiell vekst beskriver en situasjon hvor størrelsen på en befolkning (eller noe annet som vokser over tid) øker proporsjonalt med sin nåværende størrelse, noe som betyr at jo større befolkningen er, desto raskere vokser den. Dette resulterer i en J-formet kurve når befolkningen plottes over tid, som vi ser i den typiske kurven for befolkningsvekst.

Likningen er en modell for denne typen vekst og antar at vekstraten $p$ er konstant over tid. Når vi løser for  $p$ gitt de faktiske befolkningsstørrelsene ved start og slutt av perioden, finner vi en gjennomsnittlig vekstrate som ville ha resultert i den observerte veksten over de 70 årene.

Når det gjelder doblingstiden for befolkningen, er dette den tiden det tar for befolkningen å vokse til det dobbelte av sin opprinnelige størrelse ved en gitt konstant vekstrate. For en eksponensiell vekstmodell kan doblingstiden 
$T_d$ beregnes ved hjelp av følgende formel, som stammer fra den generelle eksponensielle vekstligningen:

$$T_d = \frac{\ln(2)}{\ln(1+p)}$$

Her er $ln$ den naturlige logaritmen og $p$ er den årlige vekstraten vi har beregnet. Ved å bruke denne formelen kan vi estimere hvor mange år det vil ta for jordas befolkning å fordoble seg, gitt at den beregnede vekstraten opprettholdes og at andre faktorer ikke påvirker befolkningsveksten.

La oss anta at vi har beregnet $p$ fra tidligere steg. Her er Julia-koden som vil beregne doblingstiden basert på denne vekstraten


Dette vil gi oss antall år det vil ta for jordas befolkning å fordoble seg ved den beregnede vekstraten.

```julia
doblingstid = log(2) / log(1 + p)

println("Tiden det tar for befolkningen å fordoble seg ved vekstrate p:")
println(doblingstid)

# Avrundet
println("Tiden det tar for befolkningen å fordoble seg ved vekstrate p:")
println(round(doblingstid))
```
