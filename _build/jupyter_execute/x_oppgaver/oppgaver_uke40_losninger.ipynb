```julia
# Oppgave 1 

sum = 0
for i = 1:100
    sum += i
end
println("Summen av tallene fra 1 til 100 er: $sum")

# Oppgave 2

println("")
function sjekkTall(dittTall)
    if dittTall > 0
        println("$dittTall er et positivt tall.")
    elseif dittTall < 0
        println("$dittTall er et negativt tall.")
    else
        println("$dittTall er null.")
    end
end

# Eksempel på bruk:
sjekkTall(5)
sjekkTall(-3)
sjekkTall(0)
println("")
# Oppgave 3

function summerTilN(n)
    sum = 0
    i = 1
    while i <= n
        sum += i
        if sum > 100
            break
        end
        i += 1
    end
    return sum
end

# Eksempel på bruk:
println("Summen er: ", summerTilN(20))
```

## Oppgave 7-9:

```julia
# Oppgave 7
using Plots

function lineærFunksjon(x)
    return 2x + 1
end

x_verdier = -10:10
y_verdier = lineærFunksjon.(x_verdier)
plot(x_verdier, y_verdier, label="y = 2x + 1")
```

```julia
# Oppgave 8
function kvadratiskFunksjon(x)
    return x^2 - 4x + 4
end

x_verdier = -2:0.1:6
y_verdier = kvadratiskFunksjon.(x_verdier)
plot(x_verdier, y_verdier, label="y = x^2 - 4x + 4")
```

```julia
# Oppgave 9
x_verdier = -2π:0.1:2π

# Sinusfunksjon
y_sin = sin.(x_verdier)
plot(x_verdier, y_sin, label="y = sin(x)")

# Eksponentiell funksjon
y_exp = exp.(x_verdier)
plot!(x_verdier, y_exp, label="y = e^x")
```

```julia
# Bemerk at siden exp funksjonen er så stor, så blir sinusfunksjonen nesten usynlig.
# Vi kan løse dette ved å endre på skalaen på y-aksen:

plot(x_verdier, y_sin, label="y = sin(x)", ylim=(0,10))
plot!(x_verdier, y_exp, label="y = e^x")
```

## Oppgave 10-12:

```julia
# Oppgave 10 & 12
using Plots

function eksponentiell_vekst(P, r, t)
    return P * (1 + r)^t
end

P = 1000
r = 0.05
tidsintervall = 0:1:20  # fra 0 til 20 år, i steg av 1 år
verdier = [eksponentiell_vekst(P, r, t) for t in tidsintervall]
verdier_5_01 = [eksponentiell_vekst(P, 0.1, t) for t in tidsintervall]
verdier_5_02 = [eksponentiell_vekst(P, 0.2, t) for t in tidsintervall]

plot(tidsintervall, verdier, label="Eksponentiell Vekst", xlabel="Tid (år)", ylabel="Beløp", title="Eksponentiell Vekst over Tid")
plot!(tidsintervall, verdier_5_01, label="Eksponentiell Vekst med 10% rente")
plot!(tidsintervall, verdier_5_02, label="Eksponentiell Vekst med 20% rente")


```

```julia
# Oppgave 11
eksponentiell_vekst_log_skala = log10.(verdier)
plot(tidsintervall, eksponentiell_vekst_log_skala, label="Eksponentiell Vekst", xlabel="Tid (år)", ylabel="Beløp", title="Eksponentiell Vekst over Tid (logaritmisk skala)")
```

```julia
# Oppgave 17
function rente(P,r,n)
    return P * (1 + r)^n
end

rente(1000, 0.05, 10)
```

```julia
# Oppgave 18
function beregn_npv(initial_investering, årlige_inntekter, diskonteringsrente, år)
    npv = -initial_investering
    for t in 1:år
        npv += årlige_inntekter / (1 + diskonteringsrente)^t
    end
    return npv
end


npv_prosjekt = beregn_npv(1_000_000, 300_000, 0.05, 5)
println("NPV for prosjektet er: $npv_prosjekt")


# Hvis npv_prosjekt er positiv, indikerer det at prosjektet vil gi en positiv avkastning etter å ha tatt hensyn til tidsverdien av penger, og det vil være en god investering.
# Hvis npv_prosjekt er negativ, indikerer det at prosjektet ikke vil dekke kostnadene når man tar hensyn til tidsverdien av penger, og det vil ikke være en god investering.


npv_med_økt_inntekt = beregn_npv(1_000_000, 400_000, 0.05, 5)
println("NPV for prosjektet med økt inntekt er: $npv_med_økt_inntekt")
```
