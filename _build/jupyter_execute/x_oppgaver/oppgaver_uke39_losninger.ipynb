```julia
# Oppgave 1: Funksjonsplott
function plot_functions()
    x = -10:0.1:10

    y1 = 1 ./ x
    plot(x, y1, linestyle=:dash, color=:blue, label="y = 1/x", grid=:both)
    xlabel!("x")
    ylabel!("y")
    title!("Funksjonsplott")

    y2 = sin.(x) .* cos.(x)
    plot!(x, y2, linestyle=:dot, color=:red, label="y = sin(x)cos(x)")

    y3 = 2 .* x.^2 .- 3 .* x .+ 1
    scatter!(x, y3, marker=:x, color=:red, label="y = 2x^2 - 3x + 1")
end
```

```julia
plot_functions()
```

```julia
# Oppgave 3: If-else forstÃ¥else og funksjoner
function days_in_month(month::String)
    if month == "Januar" # Kunne endret til Ã¥ bruke lowercase(month) == "januar" , som ville vÃ¦rt en bedre lÃ¸sning
        return 31
    elseif month == "Februar"
        return "28 eller 29"
    elseif month in ["Mars", "Mai", "Juli", "August", "Oktober", "Desember"]
        return 31
    elseif month in ["April", "Juni", "September", "November"]
        return 30
    else
        return "Ugyldig mÃ¥ned"
    end
end

function maximum_of_three(a, b, c)
    if a >= b && a >= c
        return a
    elseif b >= a && b >= c
        return b
    else
        return c
    end
end

function quiz_game()
    correct_answers = 0
    wrong_answers = 0

    println("SpÃ¸rsmÃ¥l 1: Er jorden flat? (Ja/Nei)")
    answer1 = readline()
    if lowercase(answer1) == "nei"
        correct_answers += 1
        println("Riktig!")
    else
        wrong_answers += 1
        println("Galt!")
    end

    println("Antall riktige svar: ", correct_answers)
    println("Antall gale svar: ", wrong_answers)
end
```

```julia
function fibonacci_numbers(N)
    a, b = 0, 1
    i = 1
    while i <= N
        println(a)
        a, b = b, a + b
        i += 1
    end
end

function gjett_tallet(max_tall)
    # Genererer et tilfeldig tall mellom 0 og max_tall
    fortsett = true
    tilfeldig_tall = rand(0:max_tall)
    
    println("Jeg tenker pÃ¥ et tall mellom 0 og $max_tall. Kan du gjette hva det er?")
    
    # Initialiserer en variabel for Ã¥ holde styr pÃ¥ antall forsÃ¸k
    antall_forsÃ¸k = 0
    

    while fortsett # Starter en uendelig lÃ¸kke
        # Ãker antall_forsÃ¸k med 1 for hver iterasjon
        antall_forsÃ¸k += 1
        
        # Ber brukeren om Ã¥ legge inn et gjetning
        print("Din gjetning:")
        gjetning = parse(Int, readline())  # Konverterer inndata til heltall
        
        # Sjekker om gjetningen er korrekt
        if gjetning == tilfeldig_tall
            println("Gratulerer! Du gjettet riktig etter $antall_forsÃ¸k forsÃ¸k.")
            break  # Avslutter lÃ¸kken
        elseif gjetning < tilfeldig_tall
            println("For lavt! PrÃ¸v igjen.")
        else
            println("For hÃ¸yt! PrÃ¸v igjen.")
        end

        # Sjekker om brukeren vil fortsette Ã¥ spille
        print("Vil du spille en ny runde? (ja/nei) ") # sjekk med bruk av lowercase
        fortsett = lowercase(readline()) == "ja" # hva skjer hvis du skriver noe annet nei? 

    end
end

function countdown(N)
    for i in N:-1:1
        println(i)
    end
    println("Blast off!")
end
```
