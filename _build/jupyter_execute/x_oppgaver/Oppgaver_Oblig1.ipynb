```
julia> collatz_steps(6)
# Output: "It takes 8 steps to reach 1 from 6."

julia> collatz_steps(27)
# Output: "It takes 111 steps to reach 1 from 27."

julia> collatz_steps(1)
# Output: "It takes 0 steps to reach 1 from 1."
```

## Oppgave 2: Aksjemarkedssimulering


**Bakgrunn:**

I den virkelige verden handler økonomi ofte om å forutsi ulike scenarier basert på tilgjengelige data. En av måtene å gjøre dette på er ved å simulere hvordan et visst økonomisk system, som for eksempel aksjemarkedet, oppfører seg over tid.

**Målet med oppgaven:**

* Lære å generere tilfeldige data
* Bruke grunnleggende kontrollstrukturer (løkker, betingelser)
* Lære å lagre og hente data fra en array eller en liste
* Visualisering av data
* Enkel bruk av funksjoner

### Oppgavebeskrivelse:

1. Generere aksjepriser: Lag en funksjon som genererer tilfeldige aksjepriser for en gitt periode (f.eks., 30 dager). Du kan starte med en initiell aksjepris og legge til tilfeldige endringer for hver dag.

2. Beregning av gjennomsnitt og volatilitet: Lag funksjoner som beregner gjennomsnittlig aksjepris og volatilitet (standardavvik) for perioden.

3. Investeringssimulering: Simuler en enkel investeringsstrategi. For eksempel, kjøp aksjen hvis den går ned 3 dager på rad og selg hvis den går opp 3 dager på rad. Se hvor mye "penger" du ender opp med etter 30 dager.

4. Visualisering: Bruk innebygde funksjoner i Julia for å visualisere aksjeprisene og investeringsstrategien over tid.

5. Rapport: Skriv en kort rapport som forklarer hva koden din gjør, og hva du har lært fra denne simuleringen.

**Tips:**

Bruk standardbiblioteker for statistikk og plotting, som `Random`, `Statistics` og `Plots`


## Oppgave 3: Investering og Avkasting

**Bakgrunn:**

Du er en ung investor som ønsker å forstå hvordan ulike typer investeringer vokser over tid. Du har bestemt deg for å lage en enkel simuleringsmodell i Julia som bruker noen grunnleggende økonomiske prinsipper.

### Oppgavebeskrivelse:

1. **Fast Rente Investering**: Skriv en funksjon i Julia som beregner endelig beløp for en investering som vokser med en fast rente over en gitt tidsperiode.

2. **Komponentrente Investering**: Lage en funksjon som simulerer vekst av en investering med  komponentrente over tid (geometrisk rekke).

3. **Investeringskalkulator**: Skriv en tredje funksjon som tar inn følgende parametre: type investering (Fast Rente eller Komponentrente), startkapital, rente, og tid. Funksjonen skal returnere det endelige beløpet av investeringen.

4. **Brukergrensesnitt og Visualisering**: Lag en funksjon som lar brukeren velge type investering, parametere og som deretter viser den endelige sluttkapitalen. Generer også en graf som viser investeringen over tid.

Koden skal være korrekt, robust og godt kommentert.
